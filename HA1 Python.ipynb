# MatPlotLib

```python
x=np.linspace(0,3,1000)
y_1=x*np.sin(10/(x+0.00001))
y_2=x/2*np.cos(10*x)
y_3=(1+x)/1.5*np.cos(11*x)
fig = plt.figure(figsize = (20,20))
plt.subplot(221)
plt.plot(x,y_1,'r', x, y_2, 'g', x, y_3, 'b')
plt.ylabel('cool variable')
plt.xlabel('another cool variable')
red_patch = mpatches.Patch(color='red', label='Red')
blue_patch = mpatches.Patch(color='blue', label='Blue')
green_patch = mpatches.Patch(color='green', label='Green')
plt.legend(handles=[red_patch,blue_patch,green_patch])
plt.show()
fig.savefig('waves.pdf')
```

# NumPy

```python
def linspice(a1, a2, n):
    a1 = float(a1)
    a2 = float(a2)
    m = (a2-a1)/(n-1)
    vect = np.zeros(n)
    vect[0]=a1
    for i in xrange(n-1):
        vect[i+1]=vect[i]+m
    return vect
```

```python
start_time = time.time()
v = linspice(1, 4, 6)
print v
print time.time() - start_time
```

```python
def linsproce(a1, a2, n):
    a1 = float(a1)
    a2 = float(a2)
    m = (a2-a1)/(n-1)
    vect = [(a1+k*m) for k in range(0, n)]
    return vect
```

```python
start_time = time.time()
v = linsproce(1,4,6)
print v
print time.time() - start_time
```

```python
start_time = time.time()
v = np.linspace(1,4,6)
print v
print time.time() - start_time
```
